```
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### 1. Scrapping the given Wikipedia link

```
wikip = "https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=945633050." #I had to use an old version of the page because the table format changed completely
fuente = requests.get(wikip).text
```

```
soup = BeautifulSoup(fuente, 'xml')
table=soup.find('table')
```

```
nombres_columnas = ['Postalcode','Borough','Neighborhood']
tabla = pd.DataFrame(columns = nombres_columnas)
```

```
for tr_cell in table.find_all('tr'):
    row_data=[]
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
    if len(row_data)==3:
        tabla.loc[len(tabla)] = row_data
```

```
tabla.head()
```

###Data Wrangling

Next step is to remove "Not assigned" elements

```
tabla=tabla[tabla['Borough']!='Not assigned']
```

```
tabla=tabla[tabla.Borough != 'Not assigned']
tabla = tabla.sort_values(by=['Postalcode','Borough'])

tabla.reset_index(inplace=True)
tabla.drop('index',axis=1,inplace=True)

tabla.head()
```

###2. Merging neighbourhoods

Neighbourhoods are grouped based on the PostalCode.

```
tabla2=tabla.groupby('Postalcode')['Neighborhood'].apply(lambda x: "%s" % ', '.join(x))
tabla2=tabla2.reset_index(drop=False)
tabla2.rename(columns={'Neighborhood':'Neighborhood_joined'},inplace=True)
```

```
tabla3 = pd.merge(tabla, tabla2, on='Postalcode')
```

```
tabla3.drop(['Neighborhood'],axis=1,inplace=True)
```

```
tabla3.rename(columns={'Neighborhood_joined':'Neighborhood'},inplace=True)
tabla3.head()
```

```
tabla3.shape #We get the shape of the table tabla3
```

#Part 3

Explore and cluster the neighborhoods in Toronto. I choosed to work with only boroughs that contain the word Toronto.

```
from geopy.geocoders import Nominatim 

address = "Toronto, ON"

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Toronto city are the following: {}, {}.'.format(latitude, longitude))
```

```
import folium # map rendering library

# create map of Toronto using latitude and longitude values
Toronto_mapa = folium.Map(location=[latitude, longitude], zoom_start=10)
Toronto_mapa
```

Next, markers are added to the map.

```
for lat, lng, borough, neighborhood in zip(
        tabla_Toronto['Latitude'], 
        tabla_Toronto['Longitude'], 
        tabla_Toronto['Borough'], 
        tabla_Toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Toronto_mapa)  

Toronto_mapa
```

Now, I will take only those boroughs that contain the word Toronto

```
Toronto_Toronto = tabla_Toronto[tabla_Toronto['Borough'].str.contains("Toronto")].reset_index(drop=True)
Toronto_Toronto.head()
```

The map is shown again with the selected boroughs

```
mapa_Torontox2 = folium.Map(location=[latitude, longitude], zoom_start=12)
for lat, lng, borough, neighborhood in zip(
        Toronto_Toronto['Latitude'], 
        Toronto_Toronto['Longitude'], 
        Toronto_Toronto['Borough'], 
        Toronto_Toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(mapa_Torontox2)  

mapa_Torontox2
```
